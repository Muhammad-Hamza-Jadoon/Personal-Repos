```python
import pandas as pd
import numpy as np
from pprint import pprint
```

```python
df = pd.read_csv("data/data_.csv")
print(df.isnull())
```

```python
df["output"].info()
```

```python
print(df[df['input'].isna()])  # Filter rows where 'A' is NaN
```

```python
# Extract authors and topics using regex
df["author"] = df["input"].str.extract(r"author:\s*([\w]+)")
df["topic"] = df["input"].str.extract(r"topic\s*([\w]+)")
```

```python
unique_authors = df["author"].nunique()
unique_topics = df["topic"].nunique()
poems_per_author = df["author"].value_counts()
poems_per_topic = df["topic"].value_counts()

# Print results
print(f"Number of unique authors: {unique_authors}")
print(f"Number of unique topics: {unique_topics}")
```

```python
print("\nPoems per author:\n", poems_per_author)
```

```python
print("\nPoems per topic:\n", poems_per_topic)
```

```python
# No of unique authors for every poem

topic_to_authors = df.groupby("topic")["author"].unique().apply(list)
print("\nTopic-Author Mapping:")
for topic, authors in topic_to_authors.items():
    # pprint(f"{topic}: {authors}")
    pprint(f"{topic}: {len(authors)}")
```

```python
# No of unique poems for every authors

author_to_topics = df.groupby("author")["topic"].unique().apply(list)
print("\nAuthor-Topic Mapping:")
for author, topics in author_to_topics.items():
    if author == "EmilyDickinson":
        pprint(f"{author}: {topics}")
        print(f"{author}: {len(topics)}")
```
