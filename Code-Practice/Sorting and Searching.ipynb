```python
# Time complexity: O(N^2)
# Space complexity: O(1)
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):

        print("\nLoop ", 0, n-i-1)
        print(arr)
        
        bool = False
        for j in range(0, n-i-1):
            print(arr[j]," > ", arr[j+1])
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                print("\nswap [",arr[j],"] with [", arr[j+1],"]")
                print("sorted array : ",arr,"\n")
                bool = True
                
        if bool == False: return arr
                
    return arr


arr = [10,2,4,3,9,8,7,5,6,1]
bubble_sort(arr)
```

```python
# Time complexity: O(N^2)
# Space complexity: O(1)
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        
        key = arr[i]
        j = i - 1
        print("\n",arr)
        # print("i = ",i, "| j = ",j, "| key = ", key, "| arr[j] = ", arr[j])

        print("comparison between : ",key, " < ", arr[j])
        while j >= 0 and key < arr[j]:
            # print()
            arr[j + 1] = arr[j]
            print("---Loop---", arr)
            j -= 1
            print("comparison between : ",key, " < ", arr[j])
        arr[j + 1] = key
        print("                                     final array : ",arr)
    return arr

arr = [1,2,4,3,9,8,7,5,6,10]
insertion_sort(arr)
```

```python
def merge(arr, l, m, r):
    # Create temporary arrays to hold the left and right subarrays
    left = arr[l:m+1]
    right = arr[m+1:r+1]

    i = j = 0  # Initial indexes of the left and right subarrays
    k = l      # Initial index of the merged subarray

    # Merge the left and right arrays back into the main array
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    # Copy any remaining elements of left, if any
    while i < len(left):
        arr[k] = left[i]
        i += 1
        k += 1

    # Copy any remaining elements of right, if any
    while j < len(right):
        arr[k] = right[j]
        j += 1
        k += 1


def iterative_merge_sort(arr):
    n = len(arr)
    # Start with subarray size 1, and keep doubling the size
    subarray_size = 1
    while subarray_size < n:
        # For each pass, merge subarrays of current size
        for left_start in range(0, n, 2*subarray_size):
            # Find the mid point and the right end for merging
            mid = min(left_start + subarray_size - 1, n-1)
            right_end = min(left_start + 2*subarray_size - 1, n-1)
            if mid < right_end:
                merge(arr, left_start, mid, right_end)
        subarray_size *= 2  # Double the size of subarrays to merge next

# Example usage
arr = [1,2,4,3,9,8,7,5,6,10]
print("Original array:", arr)
iterative_merge_sort(arr)
print("Sorted array:", arr)
```
